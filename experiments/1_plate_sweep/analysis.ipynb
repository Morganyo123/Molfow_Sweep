```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
plt.rcParams.update({
    "font.size": 11,
    "font.family": "serif",
    "axes.labelsize": 12,
    "axes.titlesize": 12,
    "legend.fontsize": 10,
    "xtick.labelsize": 10,
    "ytick.labelsize": 10,
    "figure.dpi": 150,
    "savefig.dpi": 300,
})
```

```python
df = pd.read_csv("position_sweep_summary_with_deflector.csv")
```

```python
df.head()
```

```python
plt.plot(df["offset"], df["plate_back_imping_rate"], marker="o",label="Back Impingement Rate")
plt.xlabel("Offset [cm]")
plt.ylabel("Impingement Rate")
plt.grid()
plt.legend()
```

```python
plt.plot(df["offset"], df["plate_front_imping_rate"], marker="o",label="Front Impingement Rate")
plt.xlabel("Offset [cm]")
plt.ylabel("Impingement Rate")
plt.grid(True)
plt.legend()
```

```python
fig, ax = plt.subplots(3, 1, figsize=(8, 6),constrained_layout=True)
ax[0].plot(df["offset"], df["plate_back_imping_rate"], marker="o", ms=4, lw=1.2, color="#1f77b4")
ax[0].set_ylabel("Back rate\n[hits/s]")
ax[0].tick_params(axis="y", which="major", labelsize=10,)
ax[0].grid(alpha=0.3)

ax[1].plot(df["offset"], df["plate_front_imping_rate"], marker="o", ms=4, lw=1.2, color="#d62728")
ax[1].set_ylabel("Front rate\n[hits/s]")
ax[1].grid(alpha=0.3, which="both")
ax[1].set_xlabel("Plate offset [cm]")

ax[2].plot(df["offset"], df["deflector_imping_rate"], marker="o", ms=4, lw=1.2, color="#2ca02c")
ax[2].set_ylabel("Deflector rate\n[hits/s]")
ax[2].grid(alpha=0.3, which="both")
ax[2].set_xlabel("Plate offset [cm]")

plt.savefig("position_sweep_impingement_rates.png", dpi=300, bbox_inches="tight")
```
