```python
dist = []

plate_dist = 0
cone_dist = 0


#construct plate and distance pattern
for i in range(11):
    cone_dist = plate_dist - 0.25
    for i in range(5):
        dist.append([plate_dist, cone_dist])
        cone_dist += 0.25 

    cone_dist = -0.25
    plate_dist += 0.5

print(dist)
```

```python
intensity = df['plate_back_imping_rate'] 
x = []
y = []


for pair in dist:
    x.append(pair[0])
    y.append(pair[1])
```

```python
plt.tricontourf(x, y, intensity, levels=500)
plt.xlabel("plate distance (cm)")
plt.ylabel("cone distance (cm)")
plt.title("Intensity Distribution")
plt.colorbar(label="Intensity")
plt.show();
```

```python
dist = np.array(dist)
intensity = np.array(intensity)

cone = dist[:, 0]
plate = dist[:, 1]
offset = np.round(plate - cone, 2)

fig, ax = plt.subplots()

for o in np.unique(offset):
    mask = offset == o
    order = np.argsort(plate[mask])
    ax.plot(plate[mask][order], intensity[mask][order], 'x-', label=f'offset = {o}')

ax.set_xlabel('Plate distance')
ax.set_ylabel('Intensity')
ax.legend(title='Offset (plate − cone)')
plt.show()
```
