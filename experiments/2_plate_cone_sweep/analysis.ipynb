```python
dist = []

plate_dist = 0
cone_dist = 0


#construct plate and distance pattern
for i in range(11):
    cone_dist = plate_dist - 0.25
    for i in range(5):
        dist.append([plate_dist, cone_dist])
        cone_dist += 0.25 

    cone_dist = -0.25
    plate_dist += 0.5

print(dist)
```

```python
intensity = df['plate_back_imping_rate'] 
x = []
y = []


for pair in dist:
    x.append(pair[0])
    y.append(pair[1])
```

```python
plt.tricontourf(x, y, intensity, levels=500)
plt.xlabel("plate distance (cm)")
plt.ylabel("cone distance (cm)")
plt.title("Intensity Distribution")
plt.colorbar(label="Intensity")
plt.show();
```

```python
dist = np.array(dist)
intensity = np.array(intensity)

plate = dist[:, 0]
cone = dist[:, 1]
offset = np.round(cone - plate, 2)

fig, ax = plt.subplots()

for o in np.unique(offset):
    mask = offset == o
    order = np.argsort(plate[mask])
    ax.plot(plate[mask][order], intensity[mask][order], 'x-', label=f'{o}')

ax.set_xlabel('Plate distance [cm]')
ax.set_ylabel('Impingement Rate [hits/s]')
ax.legend(title='Cone offset [cm]',fontsize=8)
ax.grid()
plt.show()

```

## Add Errors?

Assume error is related to poisson error on number of hits on the back plate. $1/\sqrt{N}$

```python
fig, ax = plt.subplots()

for o in np.unique(offset):
    mask = offset == o
    order = np.argsort(plate[mask])
    ax.errorbar(plate[mask][order], intensity[mask][order], yerr=intensity[mask][order]/np.sqrt(df["plate_back_mc_hits"][mask]), fmt="x-", label=f'{o}',capsize=3)



mask = df2['offset'] <= 5
ax.errorbar(df2['offset'][mask], df2["plate_back_imping_rate"][mask], yerr=df2["plate_back_imping_rate"][mask]/np.sqrt(df2["plate_back_mc_hits"][mask]), fmt="x-", label="No Cone", color="black",capsize=3)

ax.set_xlabel('Plate distance [cm]')
ax.set_ylabel('Impingement Rate [hits/s]')
ax.legend(title='Cone offset [cm]',loc="best", fontsize=8)
ax.grid()
plt.savefig("cone_plate_sweep.png", dpi=300, bbox_inches='tight')
```
